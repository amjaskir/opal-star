```python
import numpy as np
import matplotlib.pyplot as plt
import random
import dynamics
import plot_dynamics
```

# Original OpAL Model

```python
# initialize parameters
n_trials = 100
n_states = 100
r_mag = 1
l_mag = 0
v0 = 0.5
C = 0

alpha_c = 0.05
alpha_a = .1
beta = 10 # needed for Act figure
params = (alpha_c, alpha_a, beta)

hebb = True
crit = "SA"
norm = False
anneal = False
T = 10
```

```python
fig1, ax1 = plt.subplots(figsize=(7, 5))   # go actor evolution
fig2, ax2 = plt.subplots(figsize=(7, 5))   # no go actor evolution
fig3, ax3 = plt.subplots(figsize=(7, 5))   # act val evolution
fig4, ax4 = plt.subplots(figsize=(7, 5))   # critic evolution
axs = [ax1,ax2,ax3]

probs = np.arange(0.1,1,.1)  # probabilities to try
all_states = []
for p in probs:
    env = "%d_10_1" %(p*100)
    states = dynamics.simulate_learning(params,n_states,n_trials,env=env,v0=v0,r_mag=r_mag,l_mag=l_mag,norm=norm,crit=crit,hebb=hebb)
    plot_dynamics.avg_qs_dynamics(states,n_trials,n_states,C,axs,p,beta=1.5,rho=0.0)
    plot_dynamics.avg_vs(states,n_trials,n_states,"SA",C,ax4,p)
    all_states.append(states)
```
