Shows evolution of actor and critic weights for a variety of probabilities of reward. 

```python
import numpy as np
import matplotlib.pyplot as plt
from bogacz import Bogacz
import dynamicsb
import plot_dynamicsb
```

```python
# initialize parameters
n_trials = 100
n_states = 100
r_mag = 1           # changing either mag to 0, no real learning in the respective weight
l_mag = -1          # weird behavior on N when l_mag is 0, but val is fine
C = 0
policy="forced"
alpha_a = 0.3       # params from fig 5
epsilon= 0.22       # set epsilon to 1 and lbda to 0 for q-learner
lbda= 0.12 
params = (alpha_a, epsilon, lbda)
```

```python
fig1, ax1 = plt.subplots(figsize=(7, 5))   # go actor evolution
fig2, ax2 = plt.subplots(figsize=(7, 5))   # no go actor evolution
fig3, ax3 = plt.subplots(figsize=(7, 5))   # value evolution
axs = [ax1,ax2,ax3]

probs = np.arange(0.1,1,.1)  # probabilities to try
for p in probs:

    states = dynamicsb.simulatebogacz(params,n_states,n_trials,p=p,r_mag=r_mag,l_mag=l_mag)
    plot_dynamicsb.avg_qs(states,n_trials,n_states,C,axs,p)
```
