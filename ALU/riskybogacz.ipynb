```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import risk_bogacz

from bogacz import Bogacz
import plot_dynamicsb
```

```python

alpha_a =0.95
epsilon=0.1
K=.1
lbda=0.07
params = (alpha_a, epsilon, K, lbda)
n_states = 1000
n_trials = 40

crit = "S"
plot_all = True
plt.rcParams.update({'font.size': 22})

```

```python
env = "low"
rho=0.25
rnd_seed = random.randint(1,100000)      # use same random seed to limit variance
H_states_zero = risk_bogacz.simulatebgamble(params,n_states,n_trials,crit=crit,env=env, rho=rho,rnd_seed=rnd_seed)

# plot average sm
fig, axs = plt.subplots(figsize=(6, 5))
plot_dynamicsb.avg_sm(H_states_zero,n_trials,n_states,0,axs,"black")
rho=0.15
rnd_seed = random.randint(1,100000)      # use same random seed to limit variance
H_states_zero = risk_bogacz.simulatebgamble(params,n_states,n_trials,crit=crit,env=env, rho=rho,rnd_seed=rnd_seed)

plot_dynamicsb.avg_sm(H_states_zero,n_trials,n_states,0,axs,"purple")


plt.legend(("$\\rho=0.25$", "$\\rho=0.15$"), fontsize = 15, markerscale = 50, loc = 'lower right')
plt.title("low Probability Gambles")
plt.xlabel("Trial")
plt.ylabel("$\overline{p}(80\%)$")
plt.tight_layout()
```

```python
env = "high"
rho=0.65
rnd_seed = random.randint(1,100000)      # use same random seed to limit variance
H_states_zero = risk_bogacz.simulatebgamble(params,n_states,n_trials,crit=crit,env=env,rho=rho,rnd_seed=rnd_seed)

# plot average sm
fig, axs = plt.subplots(figsize=(6, 5))
plot_dynamicsb.avg_sm(H_states_zero,n_trials,n_states,0,axs,"black")

rho=0.85

H_states_zero = risk_bogacz.simulatebgamble(params,n_states,n_trials,crit=crit,env=env,rho=rho,rnd_seed=rnd_seed)

plot_dynamicsb.avg_sm(H_states_zero,n_trials,n_states,0,axs,"purple")

plt.legend(("$\\rho=0.65$", "$\\rho=0.85$"), fontsize = 15, markerscale = 50, loc = 'lower right')
plt.title("High Probability Gambles")
plt.xlabel("Trial")
plt.ylabel("$\overline{p}(80\%)$")
plt.tight_layout()
```
